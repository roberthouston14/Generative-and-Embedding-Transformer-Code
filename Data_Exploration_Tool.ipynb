```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the CSV file into a Pandas DataFrame
file_path = input("Please enter the file path of your CSV file: ")
df = pd.read_csv(file_path)

# Prompt user for column name of the text data
text_col_name = input("Please enter the name of the text data column: ")

# Prompt user for the column name of the binary labels
label_col_name = input("Please enter the name of the binary label column (1 represents the category of interest): ")

# Count the number of words in each text data
df["num_words"] = df[text_col_name].apply(lambda x: len(x.split()))

# Subset the data by class
ham_data = df[df[label_col_name] == 0]
spam_data = df[df[label_col_name] == 1]

# Create a box plot using seaborn
fig, ax = plt.subplots(figsize=(8,6))
sns.boxplot(x=label_col_name, y="num_words", data=df)
plt.xlabel("Class", fontsize=16)
plt.ylabel("Number of Words", fontsize=16)
plt.title("Distribution of Words per Text Data", fontsize=18)
plt.show()

# Overlay word frequency distributions
fig, ax = plt.subplots(figsize=(8,6))
sns.kdeplot(ham_data["num_words"], label="ham")
sns.kdeplot(spam_data["num_words"], label="spam")
plt.xlabel("Number of Words", fontsize=16)
plt.ylabel("Frequency", fontsize=16)
plt.title("Word Frequency Distribution by Class in " + file_path.split('/')[-1], fontsize=18)
plt.legend(fontsize=14)
plt.show()

# Show word frequency distribution by class in separate histograms
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 6))
sns.histplot(ham_data["num_words"], bins=50, ax=ax1, color="#4CB391")
ax1.set(title="Ham Text Data", xlabel="Number of Word Tokens", ylabel="Frequency")
sns.histplot(spam_data["num_words"], bins=50, ax=ax2, color="#E36F5A")
ax2.set(title="Spam Text Data", xlabel="Number of Word Tokens", ylabel="Frequency")
plt.suptitle("Word Token Count Distribution by Class in " + file_path.split('/')[-1], fontsize=18)
plt.show()

# Count the number of rows in each label category
num_1 = df[label_col_name].sum()
num_0 = len(df) - num_1

# Generate summary table
grouped = pd.DataFrame({
    "Dataset": [file_path.split('/')[-1]],
    "Class 0 Count": [num_0],
    "Class 1 Count": [num_1],
    "Ratio of Class 1 Entries": [num_1 / len(df)],
})
print(tabulate(grouped, headers='keys', tablefmt='fancy_grid'))
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from tabulate import tabulate

# Load the CSV file into a Pandas DataFrame
file_path = input("Please enter the file path of your CSV file: ")
df = pd.read_csv(file_path)

# Prompt user for column name of the text data
text_col_name = input("Please enter the name of the text data column: ")

# Prompt user for the column name of the label
label_col_name = input("Please enter the name of the label column: ")

# Prompt user for the category names
category_0 = input("Please enter the name for category 0 (e.g., ham): ")
category_1 = input("Please enter the name for category 1 (e.g., spam): ")
category_2 = input("Please enter the name for category 2 (e.g., Smishing): ")

# Convert all the labels in the label column to lowercase
df[label_col_name] = df[label_col_name].str.lower()

# Map text labels to numerical values
label_mapping = {category_0.lower(): 0, category_1.lower(): 1, category_2.lower(): 2}
df[label_col_name] = df[label_col_name].map(label_mapping)

# Count the number of words in each text data
df["num_words"] = df[text_col_name].apply(lambda x: len(x.split()))

# Subset the data by class
ham_data = df[df[label_col_name] == 0]
spam_data = df[df[label_col_name] == 1]
phishing_data = df[df[label_col_name] == 2]

# Create a box plot using seaborn
fig, ax = plt.subplots(figsize=(8,6))
sns.boxplot(x=label_col_name, y="num_words", data=df)
plt.xlabel("Class", fontsize=16)
plt.ylabel("Number of Words", fontsize=16)
plt.title("Distribution of Words per Text Data", fontsize=18)
ax.set_xticklabels([category_0, category_1, category_2])
plt.show()

# Overlay word frequency distributions
fig, ax = plt.subplots(figsize=(8,6))
sns.kdeplot(ham_data["num_words"], label="ham")
sns.kdeplot(spam_data["num_words"], label="spam")
sns.kdeplot(phishing_data["num_words"], label="phishing")
plt.xlabel("Number of Words", fontsize=16)
plt.ylabel("Frequency", fontsize=16)
plt.title("Word Frequency Distribution by Class in " + file_path.split('/')[-1], fontsize=18)
plt.legend(fontsize=14)
plt.show()

# Show word frequency distribution by class in separate histograms
fig, (ax1, ax2, ax3) = plt.subplots(1, 3, figsize=(18, 6))
sns.histplot(ham_data["num_words"], bins=50, ax=ax1, color="#4CB391")
ax1.set(title="Ham Text Data", xlabel="Number of Word Tokens", ylabel="Frequency")
sns.histplot(spam_data["num_words"], bins=50, ax=ax2, color="#E36F5A")
ax2.set(title="Spam Text Data", xlabel="Number of Word Tokens", ylabel="Frequency")
sns.histplot(phishing_data["num_words"], bins=50, ax=ax3, color="#2D98DA")
ax3.set(title="Phishing Text Data", xlabel="Number of Word Tokens", ylabel="Frequency")
plt.suptitle("Word Token Count Distribution by Class in " + file_path.split('/')[-1], fontsize=18)
plt.show()

# Count the number of rows in each label category
num_0 = (df[label_col_name] == 0).sum()
num_1 = (df[label_col_name] == 1).sum()
num_2 = (df[label_col_name] == 2).sum()

# Add a column chart to show the relative count of each category
fig, ax = plt.subplots(figsize=(8, 6))
sns.countplot(x=label_col_name, data=df)
plt.xlabel("Class", fontsize=16)
plt.ylabel("Count", fontsize=16)
plt.title("Number of Samples per Category", fontsize=18)
ax.set_xticklabels([category_0, category_1, category_2])
plt.show()

# Calculate the total number of valid label rows
total_valid_rows = num_0 + num_1 + num_2

# Generate summary table
grouped = pd.DataFrame({
    "Dataset": [file_path.split('/')[-1]],
    "Class 0 ({}):".format(category_0): [num_0],
    "Class 1 ({}):".format(category_1): [num_1],
    "Class 2 ({}):".format(category_2): [num_2],
    "Ratio of Class 0 Entries": [num_0 / total_valid_rows],
    "Ratio of Class 1 Entries": [num_1 / total_valid_rows],
    "Ratio of Class 2 Entries": [num_2 / total_valid_rows],
})
print(tabulate(grouped, headers='keys', tablefmt='fancy_grid'))
```
