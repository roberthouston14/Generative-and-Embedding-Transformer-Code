```python
import numpy as np
import pandas as pd
import seaborn as sns
from sklearn.preprocessing import LabelEncoder, label_binarize
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score, roc_curve, auc
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import AdaBoostClassifier, RandomForestClassifier, ExtraTreesClassifier
from sklearn.svm import SVC, LinearSVC
from sklearn.linear_model import LogisticRegression, SGDClassifier
from sklearn.naive_bayes import GaussianNB, MultinomialNB
from sklearn.neighbors import KNeighborsClassifier
from xgboost import XGBClassifier
from sklearn.ensemble import GradientBoostingClassifier
from xgboost import XGBClassifier
from lightgbm import LGBMClassifier
from catboost import CatBoostClassifier
from sklearn.metrics import roc_curve, auc
from sklearn.preprocessing import label_binarize
import seaborn as sns
import os

# Prompt user for paths to train and test datasets
train_data_path = input("Enter the path to the training data: ")
test_data_path = input("Enter the path to the testing data: ")

# Get file names from paths
train_file_name = os.path.basename(train_data_path)
test_file_name = os.path.basename(test_data_path)

# Load train and test data
train_data_df = pd.read_csv(train_data_path)
test_data_df = pd.read_csv(test_data_path)

# Prompt user for column names
text_column = input("Enter the column name for the text data: ")
label_column = input("Enter the column name for the labels: ")

# Drop NaNs
train_data_df.dropna(subset=[text_column, label_column], inplace=True)
test_data_df.dropna(subset=[text_column, label_column], inplace=True)

# Prepare data (already included in your original code)
X_train = train_data_df[text_column]
y_train = train_data_df[label_column]
X_test = test_data_df[text_column]
y_test = test_data_df[label_column]

# Encode labels
label_encoder = LabelEncoder()
y_train = label_encoder.fit_transform(y_train)
y_test = label_encoder.transform(y_test)

# Vectorize the text data
vectorizer = TfidfVectorizer()
X_train = vectorizer.fit_transform(X_train)
X_test = vectorizer.transform(X_test)

# Define models
models = [
    ("SGDClassifier", SGDClassifier(loss='modified_huber')),
    ("AdaBoost", AdaBoostClassifier(base_estimator=DecisionTreeClassifier(max_depth=3), n_estimators=100)),
    ("Random Forest", RandomForestClassifier()),
    ("Extra Trees", ExtraTreesClassifier()),
    ("SVC", SVC(probability=True)),
    ("Logistic Regression", LogisticRegression(max_iter=1000)),
    ("Multinomial Naive Bayes", MultinomialNB()),
    ("Gradient Boosting", GradientBoostingClassifier()),
    ("XGBoost", XGBClassifier(use_label_encoder=False, eval_metric='mlogloss')),
    ("LightGBM", LGBMClassifier()),
    ("CatBoost", CatBoostClassifier(silent=True)),
]

# Run models and generate metrics and plots
metrics = []

for name, model in models:
    print(f"Running {name}...")

    # Check if the model requires dense input data
    if name == 'Gaussian Naive Bayes':
        model.fit(X_train.toarray(), y_train)
        y_pred = model.predict(X_test.toarray())
        y_prob = model.predict_proba(X_test.toarray()) if hasattr(model, 'predict_proba') else None
    else:
        model.fit(X_train, y_train)
        y_pred = model.predict(X_test)
        y_prob = model.predict_proba(X_test) if hasattr(model, 'predict_proba') else None

    print(f"\n{name} Classification Report:")
    cr = classification_report(y_test, y_pred, target_names=label_encoder.classes_, output_dict=True)
    print(classification_report(y_test, y_pred, target_names=label_encoder.classes_))

    print(f"\n{name} Confusion Matrix:")
    fig, ax = plt.subplots(figsize=(8, 6))
    cm = confusion_matrix(y_test, y_pred)
    sns.heatmap(cm, annot=True, fmt="d", cmap="YlGnBu", xticklabels=label_encoder.classes_, yticklabels=label_encoder.classes_, ax=ax)
    ax.set_xlabel("Predicted")
    ax.set_ylabel("True")
    ax.set_title(f"Confusion Matrix ({name} - Train Data: {train_file_name})")
    plt.show()

    # Calculate F1-scores for each class
    cr = classification_report(y_test, y_pred, output_dict=True)
    f1_ham = cr["0"]["f1-score"]
    f1_spam = cr["1"]["f1-score"]
    f1_smishing = cr["2"]["f1-score"]

    # Assign weights to the F1-scores based on their importance
    weight_ham = 1
    weight_spam = 2
    weight_smishing = 3

    # Calculate the custom aggregate score
    aggregate_score = ((f1_ham * weight_ham) + (f1_spam * weight_spam) + (f1_smishing * weight_smishing)) / (weight_ham + weight_spam + weight_smishing)

    print(f"\n{name} Aggregate Score: {aggregate_score:.2f}\n")

    metrics_dict = {
        "Model": name,
        "Accuracy": accuracy_score(y_test, y_pred),
        "Classification Report": cr,
        "Confusion Matrix": cm,
        "ROC Curve": {
            "fpr": None,
            "tpr": None,
            "roc_auc": None
        }
    }

    if y_prob is not None:
        y_true_bin = label_binarize(y_test, classes=range(len(label_encoder.classes_)))
        n_classes = y_true_bin.shape[1]
        fpr, tpr, _ = roc_curve(y_true_bin.ravel(), y_prob.ravel())
        roc_auc = auc(fpr, tpr)

        metrics_dict["ROC Curve"]["fpr"] = fpr
        metrics_dict["ROC Curve"]["tpr"] = tpr
        metrics_dict["ROC Curve"]["roc_auc"] = roc_auc

    metrics.append(metrics_dict)

# Generate aggregate ROC curve
plt.figure(figsize=(10, 8))
plt.plot([0, 1], [0, 1], linestyle="--", color="gray")
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.title(f"Aggregate ROC Curve (Train Data: {train_file_name})")

for metric in metrics:
    if metric["ROC Curve"]["fpr"] is not None:
        name = metric["Model"]
        fpr = metric["ROC Curve"]["fpr"]
        tpr = metric["ROC Curve"]["tpr"]
        roc_auc = metric["ROC Curve"]["roc_auc"]
        plt.plot(fpr, tpr, label=f"{name} (AUC = {roc_auc:.2f})")

plt.legend(loc="lower right")
plt.show()

# Process metrics for a more readable table
processed_metrics = []

for metric in metrics:
    new_metric = {
        "Model": metric["Model"],
        "Accuracy": metric["Accuracy"],
    }

    for label, value in metric["Classification Report"].items():
        if label not in ['accuracy', 'macro avg', 'weighted avg']:
            category_name = label_encoder.inverse_transform([int(label)])[0]
            new_metric[f"{category_name} Precision"] = value['precision']
            new_metric[f"{category_name} Recall"] = value['recall']
            new_metric[f"{category_name} F1-score"] = value['f1-score']

    # Calculate F1-scores for each class
    f1_ham = metric["Classification Report"]["0"]["f1-score"]
    f1_spam = metric["Classification Report"]["1"]["f1-score"]
    f1_smishing = metric["Classification Report"]["2"]["f1-score"]

    # Add the Aggregate Score to the new_metric dictionary
    aggregate_score = ((f1_ham * weight_ham) + (f1_spam * weight_spam) + (f1_smishing * weight_smishing)) / (weight_ham + weight_spam + weight_smishing)
    new_metric["Aggregate Score"] = aggregate_score

    processed_metrics.append(new_metric)

# Display metrics table
processed_metrics_df = pd.DataFrame(processed_metrics)
pd.set_option('display.max_colwidth', None)
pd.set_option('display.expand_frame_repr', False)
print(f"\nMetrics Table (Train Data: {train_file_name}):")
print(processed_metrics_df.to_string(index=False))

# Get the directory and base file name of the training file
train_dir, train_base_name = os.path.split(train_data_path)
train_base_name_no_ext = os.path.splitext(train_base_name)[0]

# Save the .csv file in the same directory with the same name as the training file but with a _table.csv extension
output_csv_path = os.path.join(train_dir, f"{train_base_name_no_ext}_table.csv")
processed_metrics_df.to_csv(output_csv_path, index=False)
print(f"Metrics table saved as '{output_csv_path}'.")
```
